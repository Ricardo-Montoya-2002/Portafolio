## <font color=#33A5FF>****Análisis de Datos English Premier League (EPL) 2020-2021****<tfont>

```python
#Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Cargamos los **datos** para luego ***limpiar nuestros datos***.

```python
#cargamos un archivo CSV separado por comas
epl = pd.read_csv('EPL_20_21.csv')
epl
```

```python
#leer 5 primeras lineas
epl.head()
```

```python
#podemos obtener la cantidad de filas y columnas con shape
epl.shape
```

```python
#podemos obtemer la cantidad de  filas con la funcion  len
len(epl)
```

```python
#podemos obetener la cantidad de lineas usando la ubicacion de los elementos del comando shape 
epl.shape[0]
```

```python
#leer 4 primeras lineas aleatorias
epl.sample(4)
```

```python
#Información del DataFrame
epl.info()
```

```python
#Resumen de Datos estadisticos ---- Solo columnas numéricas
epl.describe().T
```

```python
#Número total de valores nulos
epl.isna().sum()
```

```python
epl.columns
```

Creamos 2 columnas mas **Minutos por partido** y ***Goles por partido***.

```python
# creacion de 2 columnas
epl['minpermatch']=epl['Mins']/epl['Matches']
epl['goalspermatch']=epl['Goals']/epl['Matches']
```

```python
#visualizamos las nuevas columnas
epl.head()
```

```python
#onservamos la nueva cantidad de columnas de nuestro dataframe
epl.shape
```

```python
#Total de goles ---- temporada EPL 2020-2021 ultima temporada
total_goals=epl['Goals'].sum()
total_goals
```

```python
#Goles x Penales ---- temporada EPL 2020-2021 ultima temporada
total_penalty_goals=epl['Penalty_Goals'].sum()
total_penalty_goals
```

```python
epl.columns
```

```python
# Total Penales ---- temporada EPL 2020-2021 ultima temporada
total_penaltys=epl['Penalty_Attempted'].sum()
total_penaltys
```

### Gráfico Pie chart **Penales Marcados** y **Penales Atajados**.

```python
# Gráficos de penales
plt.figure(figsize=(13,6))
pnlt_not_scored=total_penaltys-total_penalty_goals
data=[pnlt_not_scored,total_penalty_goals]
nom=["Penales marcados","penales atajados"]
color=sns.color_palette('Set2')
plt.pie(data,labels=nom,colors=color,autopct='%.0f%%')
plt.show()
```

```python
epl.columns
```

```python
# Posiciones únicas de cada jugador 
# Portero (GK), defensor (DF), centrocampista (MF) delantero (FW)
epl['Position'].unique()
```

```python
# Filtrar jugadores delanteros (FW)
epl[epl['Position']=='FW']
```

```python
# Cantidad jugadores delanteros (FW)
epl_FW=epl[epl['Position']=='FW']
epl_FW['Name'].count()
```

```python
#  Nacionalidad de los Jugadores
np.size((epl['Nationality'].unique()))
```

```python
# Nacionalidad de la mayoría de los jugadores
nationality=epl.groupby('Nationality').size().sort_values(ascending=False)
#procedemos a graficarlo con bar
nationality.head(12).plot(kind='bar',figsize=(15,5),color=sns.color_palette('magma'))
plt.show()
```

```python
epl.columns
```

```python
# Los Club con maximo numero de jugadores en la temporada
epl['Club'].value_counts().nlargest(10).plot(kind='bar',
                                           figsize=(15,5),color=sns.color_palette('viridis'))
plt.show()
```

```python
# Los Club con maximo numero de jugadores en la temporada
numperclubs=epl.groupby('Club').size().sort_values(ascending=False)
numperclubs
```

```python
# Los Club con menor numero de jugadores en la temporada
epl['Club'].value_counts().nsmallest(10).plot(kind='bar',
                                           figsize=(15,5),color=sns.color_palette('viridis'))
plt.show()
```

```python
epl['Club'].unique()
```

### Edad de los jugadores

```python
epl.columns
```

```python
# Grupo de edades de los Jugadores
under20=epl[epl['Age']<=20]
age20_25=epl[(epl['Age']>20 ) & (epl['Age']<=25)]
age25_30=epl[(epl['Age']>25 ) & (epl['Age']<=30)]
up30=epl[epl['Age']>30]
```

```python
v=np.array([under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),up30['Name'].count()])
subtitle=['<=20','>20 & <=25','>25 & <=30','>30']
plt.title('Grupods de edad se los jugadores de la Premier League',fontsize=20)
plt.pie(v,labels=subtitle,autopct='%.1f%%')
plt.show()
```

```python
# Total de jugadores menores de 20 años en cada club
players_under_20=epl[epl['Age']<20]
players_under_20['Club'].value_counts().head(19).plot(kind='bar',color=sns.color_palette('cubehelix'),figsize=(14,8))
plt.show()
```

```python
# Jugadores menores de 20 años en  club Manchester United (7)
players_under_20[players_under_20['Club']=='Manchester United']
```

```python
# Jugadores menores de 20 años en  club Chelsea (2)
players_under_20[players_under_20['Club']=='Chelsea']
```

```python
# Gráfico BoxPlot de la edad promedio de los jugadores de cada club
plt.figure(figsize=(15,8))
sns.boxplot(x='Club',y='Age', data=epl)
plt.xticks(rotation=90)
plt.show()
```

```python
# Clubs con Edades promedio mas altas
num_player=epl.groupby('Club').size()
prom=(epl.groupby('Club')['Age'].sum())/num_player
prom.sort_values(ascending=False)
```

```python
epl.columns
```

```python
# Grafica de Total de asistencias gol por cada club
Assist_by_club=pd.DataFrame(epl.groupby('Club',as_index=False)['Assists'].sum())
sns.set(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=Assist_by_club.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Club',fontsize=50)
ax.set_ylabel('Assists',fontsize=50)
plt.xticks(rotation=75,fontsize=20)
plt.yticks(fontsize=20)
plt.rcParams['figure.figsize']=(25,15)
plt.title('Grafica de la cantidad asistencias por club',fontsize=40)
plt.show()
```

```python
# Jugadores top de 10 Asistencias Gol
top10_assistens=epl[['Assists','Name','Club','Matches']].nlargest(n=10,columns='Assists')
top10_assistens
```

```python
# Jugadores top de 10 Asistencias Gol
#Grafica
Assist_by_club=pd.DataFrame(epl[['Assists','Name']].nlargest(n=10,columns='Assists'))
sns.set(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Name',y='Assists',data=Assist_by_club.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Name',fontsize=50)
ax.set_ylabel('Assists',fontsize=50)
plt.xticks(rotation=75,fontsize=20)
plt.yticks(fontsize=20)

plt.rcParams['figure.figsize']=(25,15)
plt.title('Grafica de la cantidad asistencias por jugador',fontsize=40)
plt.show()
```

```python
# Grafica del total de goles de cada club
Assist_by_club=pd.DataFrame(epl.groupby('Club',as_index=False)['Goals'].sum())
sns.set(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Assist_by_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club',fontsize=50)
ax.set_ylabel('Goals',fontsize=50)
plt.xticks(rotation=75,fontsize=20)
plt.yticks(fontsize=20)
plt.rcParams['figure.figsize']=(25,15)
plt.title('Grafica de la cantidad goles por club',fontsize=40)
plt.show()
```

```python
#Top 10 de  Goles por Jugador
top10_goals=epl[['Goals','Name','Club','Matches']].nlargest(n=10,columns='Goals')
top10_goals
```

```python
#Top 10 de  Goles por Jugador
#grafica
Assist_by_club=pd.DataFrame(epl[['Goals','Name']].nlargest(n=10,columns='Goals'))
sns.set(style='whitegrid',color_codes=True)
ax=sns.barplot(x='Name',y='Goals',data=Assist_by_club.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Name',fontsize=50)
ax.set_ylabel('Goals',fontsize=50)
plt.xticks(rotation=75,fontsize=20)
plt.yticks(fontsize=20)

plt.rcParams['figure.figsize']=(25,15)
plt.title('Grafica de la cantidad goles por jugador',fontsize=40)
plt.show()
```

```python
epl.columns
```

```python
# Top 10 Goles por partido
top10_goals_permatch=epl[['goalspermatch','Name','Club','Matches']].nlargest(n=10,columns='goalspermatch')
top10_goals_permatch
```

```python
epl.columns
```

```python
# Grafica Pi Chart - Goles con asistencia y Goles sin asistencia
plt.figure(figsize=(10,8))
assists=epl['Assists'].sum()
data=[total_goals-assists,assists]
nom=['Goles sin asistencia','goles con asistencia']
colores=sns.color_palette('Set2')
plt.pie(data,labels=nom,colors=colores,autopct='%.0f%%')
plt.show()
```

```python
epl.columns
```

```python
#Top 10 jugadores con mas tarjetas amarillas
epl_yelow=epl.sort_values(by='Yellow_Cards',ascending=False)[:10]
epl_yelow    
```
